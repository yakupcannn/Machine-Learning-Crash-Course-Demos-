```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
data = pd.read_csv("hands_on_datasets/Advertising.csv",index_col=0)
```

```python
data.head()
```

```python
plt.figure(figsize=(18,6))
plt.scatter(data["TV"],data["sales"],c="r")
plt.xlabel("TV sales based on ads ($)")
plt.ylabel("Sales (k$)")
plt.show()
```

```python
X = data["TV"].values.reshape(-1,1)
y=data["sales"].values.reshape(-1,1)
lr = LinearRegression()
lr.fit(X,y)
predictions=lr.predict(X)
plt.figure(figsize=(16,8))
plt.scatter(X,y,c="r")
plt.plot(X,predictions,c="b")
plt.xlabel("TV sales based on ads ($)")
plt.ylabel("Sales (k$)")
plt.show()

```

```python
X = data["TV"]
y=data["sales"]
exog = sm.add_constant(X)
est = sm.OLS(y,exog).fit()
print(est.summary())
```

## Multiple Linear Regression

```python
X = data.drop("sales",axis=1)
y = data["sales"].values.reshape(-1,1)
mlr = LinearRegression()
mlr.fit(X,y)
```

```python
X = np.column_stack([data["TV"],data["radio"],data["newspaper"]])
y= data["sales"].values.reshape(-1,1)
exog = sm.add_constant(X)
est = sm.OLS(y,exog).fit()
print(est.summary())
```

```python
X = data.drop("sales",axis = 1 )
X = X.iloc[:,:-1]
y = data["sales"].values.reshape(-1,1)
```

```python
mlr = LinearRegression()
mlr.fit(X,y)
```

```python
X = np.column_stack([data["TV"],data["radio"]])
y= data["sales"].values.reshape(-1,1)
exog = sm.add_constant(X)
est = sm.OLS(y,exog).fit()
print(est.summary())
```
